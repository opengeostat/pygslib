#Gaussian Anamorthosis with declustering

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import pylab as plt
import pygslib as gslib
%matplotlib inline
from scipy.stats import norm
```

## Get some data ready for work and build experimental transformation

The only way to make it work with declustering data was:

- nscore [an despike] with ``gslib.nonlinear.ttable`` and get ``yt, zt, f``
- take a regular sample on probabilities from experimental ``f.min(), f.max()``
- convert ``p`` to ``y`` values using ``y = norm.ppf(p)``
- back transform ``y`` to ``z`` using ``gslib.__dist_transf.backtr`` with transformation table ``yt, zt``
- finally do gaussian anamorphosis with ``y, z``

For non declustered samples the algorithm can be simplified to: 

- nscore [an despike] with ``gslib.nonlinear.ttable`` and get ``y, z``
- finally do gaussian anamorphosis with ``y, z``

<p>
<div class="alert alert-danger">
 ** Warning: ** There may be some bias if you do anamorphosis with declustered raw data. 
</div>

<p>
<div class="alert alert-info">
 ** Info: ** By using "despikes" in the way proposed the local atoms and isolated values are well reproduced in the experimental anamorphosis. 
</div>

```python
#get the raw data in gslib format into a pandas Dataframe
mydata= gslib.read_gslib_file('../datasets/cluster.dat')

# this time no despiking
zt,yt,f,w = gslib.nonlinear.ttable (mydata['Primary'].values, 
                                  mydata['Declustering Weight'].values,
                                  despkn=0,despks=0.01, despkm=True)

#get nice homogeneus distribution
p= np.linspace(f.min(), f.max(),yt.shape[0]*100) 

y = norm.ppf(p)

z,error = gslib.__dist_transf.backtr(y,zt,yt,
                                     ltail=1,utail=1,ltpar=zt.min(),utpar=zt.max(),
                                     zmin=zt.min(),zmax=zt.max(),getrank=False)

#mydata
```

```python
# plot experimental anamorphosis function 
plt.step(y,z,'.', where='mid', label='raw or experimental')
plt.xlabel('Y: Gaussian values')
plt.ylabel('Z: Raw values')
plt.title('Experimental Gaussian anamorphosis')
#plt.yscale ('log')
#plt.xscale ('log')
#plt.legend(loc=2)
plt.show()
```

## Test anamorphosis

```python
K=50
H=gslib.nonlinear.recurrentH(y=y, K=K)
# some test plot, example lack of correlation
plt.scatter(H[1,:], H[2,:])
plt.show()
plt.scatter(H[1,:], H[3,:])
plt.show()

print H.shape, y.shape

pd.DataFrame (H.T, columns=range(H.shape[0]))
```

```python
# plot some H values
plt.plot(y,H[0,:], label='H0')
plt.plot(y,H[1,:], label='H1')
plt.plot(y,H[2,:], label='H2')
plt.plot(y,H[5,:], label='H5')
plt.legend()
plt.show()
```

## Fit Anamaorphosis coefficients

```python
mean = np.sum(mydata['Primary'].values*(mydata['Declustering Weight'].values/ \
       mydata['Declustering Weight'].values.sum()))
print mean

PCI, g=gslib.nonlinear.fit_PCI(z,y,H, meanz=mean)

```

### Validate PCI

There are two basic validations:

 - The variance calculated from PCI may be close to the experimental variance (declustered)
 - The error of the z' fitted with hermite polynomial expansion may be close to z

```python
# test variance
print 'Variance from PCI:                 ', gslib.nonlinear.var_PCI(PCI)
print 'Variance experimental:             ', z.var(), mydata['Primary'].var()
```

```python
# test the fiiting error with polynomial expansion
ze = gslib.nonlinear.expand_anamor(PCI,H,r=1) # make sure that r == 1 for point support

fit_error = z-ze

bias = fit_error.mean()

#print some comparisons
print 'Mean z', z.mean()
print 'Mean ze', ze.mean()
print 'Mean er', fit_error.mean()
print 'Var z', z.var()
print 'Var ze', ze.var()
print 'Var er', fit_error.var()
print 'Bias', bias


# scaterplots of true / estimate
plt.plot(z,ze, '+')
plt.show()

# error by y
plt.plot(y,fit_error)
plt.show()

# plot histogram of errors
pd.DataFrame({'fit_error': fit_error}).hist(bins=20)
plt.show()
```

```python
# also plot anamorphosis function 
plt.plot(y,ze,label='Z from anamorphosis')
plt.step(y,z,where='mid', label='raw')
plt.step(yt,zt,where='mid', label='original')
plt.xlabel('Y: Gaussian values')
plt.ylabel('Z: Raw values')
plt.title('Gaussian anamorphosis')
plt.legend(loc=2)
plt.show()
```

```python
#plt.plot(rz, norm.cdf(ry))
plt.step(z,norm.cdf(y), label='experimental')
plt.plot(ze, norm.cdf(y), label='fit')
#plt.xlim(xmin=0.05)
plt.axis([0.05,10.0, 0,1])
plt.xlabel('Z')
plt.ylabel('P(Z<zi)')
plt.xscale('log')
plt.legend(loc=2)
plt.show()
```

# Using the function to transform arbitrary y values

As result of some nonlinear and simulation application you may have arbitrary gaussian (Y) values you want to convert to raw (Z) values. This is partially implemented and only valid for some practical intervals. 

To do this is necessary to identify four set of control points: 

 - [yamin,zamin] admissible minimum value
 - [ypmin,zpmin] practical minimum value
 - [ypmax,zpmax] practical minimum value
 - [yamax,zamin] admissible minimum value





```python
zamin, yamin, zpmin, ypmin, zpmax, ypmax, zamax, yamax = gslib.nonlinear.ctrpoints(z,y,ze)


print 'Minimum absolute value for Y  =', yamin
print 'Maximum absolute value for Y  =', yamax
print 'Minimum absolute value for Z  =', zamin
print 'Maximum absolute value for Z  =', zamax

print 'Minimum practical value for Y =', ypmin
print 'Maximum practical value for Y =', ypmax
print 'Minimum practical value for Z =', zpmin
print 'Maximum practical value for Z =', zpmax
```

```python
# also plot anamorphosis function 
plt.plot(y,ze,'.', label='Z from anamorphosis')
plt.step(y,z, '-', label='raw')
plt.plot(yamin,zamin, 'ok', label='Admissible control point')
plt.plot(yamax,zamax, 'ok')
plt.plot(ypmin,zpmin, '+r', label='Practical control point')
plt.plot(ypmax,zpmax, '+r')
plt.xlabel('Y: Gaussian values')
plt.ylabel('Z: Raw values')
plt.title('Gaussian anamorphosis')
plt.legend(loc=2)
plt.show()
```

```python
#test using different Y data to fit a new Z (not in the input)
yyy=np.linspace(-5,5,4000)

ZZZ=gslib.nonlinear.Y2Z(yyy,PCI,zamin, yamin, zpmin, ypmin, zpmax, ypmax, zamax, yamax, r=1)

plt.plot(yyy,ZZZ, label = 'z from y in [-5, 5]')
plt.plot(y,z, label = 'Raw transformation (z)')
plt.plot(y,ze, label = 'z from y in dataset (ze)')
plt.legend(loc=2)
plt.show()
```

```python
from scipy.stats import norm
plt.plot(z, g)
plt.plot(ZZZ, norm.pdf(yyy))
plt.xlabel('Z')
plt.ylabel('P(Z<zi)')
plt.xscale('log')
plt.show()
```

```python
#test using different Z data to fit a new Y 
zzz=np.linspace(z.min()-1,z.max()+3,1000)

YYY=gslib.nonlinear.Z2Y_linear(zzz,ze,y,zamin, yamin, zpmin, ypmin, zpmax, ypmax, zamax, yamax)

plt.plot(YYY,zzz, '.',linewidth=2.0, label = 'Y from arbitrary z')
plt.step(y,z, label = 'Raw transformation (z)')
plt.plot(y,ze, label = 'z from y in dataset (ze)')
plt.legend(loc=2)
plt.show()
```
