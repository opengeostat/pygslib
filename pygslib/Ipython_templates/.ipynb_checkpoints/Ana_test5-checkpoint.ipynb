## Get some data ready for work

This time we are generating z,y values with back transformmation of aribitrary value

```python
#get the data in gslib format into a pandas Dataframe
mydata= gslib.read_gslib_file('../datasets/cluster.dat')
transin,transout, error = gslib.__dist_transf.ns_ttable(mydata['Primary'],mydata['Declustering Weight'])
mydata['NS_Primary'] = gslib.__dist_transf.nscore(mydata['Primary'],transin,transout,getrank=False)
print 'there was any error?: ', error!=0

#y=np.linspace(-1.217341303825,2.983975887299, 1050)

y=np.linspace(mydata['NS_Primary'].min()-0.5 ,mydata['NS_Primary'].max()+0.5, 1050)

pd.DataFrame({'y':y}).to_csv('y.dat')

z,error = gslib.__dist_transf.backtr(y,transin,transout,
                                     ltail=1,utail=1,ltpar=0,utpar=60,
                                     zmin=0,zmax=60,getrank=False)
print 'there was any error?: ', error!=0, error
print mydata['NS_Primary'].min(),mydata['NS_Primary'].max()
```

# Using the function to transform arbitrary y values

As result of some nonlinear and simulation application you may have arbitrary gaussian (Y) values you want to convert to raw (Z) values. This is partially implemented and only valid for some practical intervals. 

To do this is necessary to identify four set of control points: 

 - [yamin,zamin] admissible minimum value
 - [ypmin,zpmin] practical minimum value
 - [ypmax,zpmax] practical minimum value
 - [yamax,zamin] admissible minimum value





```python
zamin, yamin, zpmin, ypmin, zpmax, ypmax, zamax, yamax = gslib.nonlinear.ctrpoints(z,y,ze)
print 'Minimum absolute value for Y  =', yamin
print 'Maximum absolute value for Y  =', yamax
print 'Minimum absolute value for Z  =', zamin
print 'Maximum absolute value for Z  =', zamax

print 'Minimum practical value for Y =', ypmin
print 'Maximum practical value for Y =', ypmax
print 'Minimum practical value for Z =', zpmin
print 'Maximum practical value for Z =', zpmax
```

```python
# also plot anamorphosis function 
plt.plot(y,ze,'.', label='Z from anamorphosis')
plt.plot(y,z, '+', label='raw')
plt.plot(yamin,zamin, 'ok', label='Admissible control point')
plt.plot(yamax,zamax, 'ok')
plt.plot(ypmin,zpmin, '+r', label='Practical control point')
plt.plot(ypmax,zpmax, '+r')
plt.xlabel('Y: Gaussian values')
plt.ylabel('Z: Raw values')
plt.title('Gaussian anamorphosis')
plt.legend(loc=2)
#plt.xlim([-2.5, -1])
plt.show()
```

# Change of support and information effect

## Support effect

To account for support effect you need to know: 

- point-support variance:  $Var(Zx) = (\infty)$
- block-support variance: $ Var(Zv) = \gamma(\infty)  \gamma(v,v) $

## Information effect

To account for information effect, in addition to the point variance and block variance, you need to know:

- estimated grade variance: $ Var(Zv^*) = Var(Zv) -  \sigma^2_K - 2\mu$
- true gradeestimated grade covariance: $Cov(Zv,Zv^*) = Var(Zv) - \sigma^2_K - \mu$

Note that the information effect has two components. The smoothing due to kriging, represented by the estimated grade variance and some degree of conditional bias represented by the true gradeestimated grade covariance. Note that the amount of conditional bias can be represented as the slope of regression: $p = Cov(Zv,Zv^*) / (Cov(Zv,Zv^*)  - \mu)$
