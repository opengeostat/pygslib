## Testing nonlinear (temporal development)

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import pylab as plt
import pygslib as gslib
%matplotlib inline
from scipy.stats import norm
```

## Get some data ready for work and build experimental transformation

```python
#get the raw data in gslib format into a pandas Dataframe
mydata= gslib.read_gslib_file('../datasets/cluster.dat')

z = np.array([1,2,3,5,6,7,8, 9.])
w = np.ones(z.shape[0])

zz, yy, ff, ww = gslib.nonlinear.ttable2(z,w)    

# compare some results with gslib
z,error = gslib.__dist_transf.backtr(yy,zz,yy,
                                     ltail=1,utail=1,ltpar=0,utpar=60,
                                     zmin=zz.min(),zmax=zz.max(),getrank=False)
print zz
print ww
print yy
print ff
```

## Get transformation table

```python
# plot experimental anamorphosis function 
plt.step(yy,zz, where='mid',label='raw or experimental')
plt.xlabel('Y: Gaussian values')
plt.ylabel('Z: Raw values')
plt.title('Experimental Gaussian anamorphosis')
#plt.yscale ('log')
#plt.xscale ('log')
#plt.legend(loc=2)
plt.show()

plt.step(yy,ff, where='mid')
plt.show()
plt.step(zz,ff, where='mid')
plt.show()
```

## Fit Anamaorphosis coefficients

```python
PCI, g=gslib.nonlinear.fit_PCI(zz,yy,H)

```

### Validate PCI

There are two basic validations:

 - The variance calculated from PCI may be close to the experimental variance (declustered)
 - The error of the z' fitted with hermite polynomial expansion may be close to z

```python
# test variance
print 'Variance from PCI:                 ', gslib.nonlinear.var_PCI(PCI)
print 'Variance experimental:             ', z.var()
```

```python
# test the fiiting error with polynomial expansion
ze = gslib.nonlinear.expand_anamor(PCI,H,r=1) # make sure that r == 1 for point support

fit_error = zz-ze

bias = fit_error[fit_error<4].mean()

#print some comparisons
print 'Mean z', zz.mean()
print 'Mean ze', ze.mean()
print 'Mean er', fit_error.mean()
print 'Var z', zz.var()
print 'Var ze', ze.var()
print 'Var er', fit_error.var()
print 'Bias', bias


# scaterplots of true / estimate
plt.plot(zz,ze, '+')
plt.show()

# error by y
plt.plot(yy,fit_error)
plt.show()

# plot histogram of errors
pd.DataFrame({'fit_error': fit_error}).hist(bins=20)
plt.show()
```

```python
# also plot anamorphosis function 
plt.plot(yy,ze, '-+',label='Z from anamorphosis')
plt.step(yy,zz, '-o', where='mid', label='raw')
plt.xlabel('Y: Gaussian values')
plt.ylabel('Z: Raw values')
plt.title('Gaussian anamorphosis')
plt.legend(loc=2)
plt.show()
```
