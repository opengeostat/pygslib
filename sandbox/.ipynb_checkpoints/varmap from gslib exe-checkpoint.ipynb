```python
import varmap
import pygslib
```

# test with arbitrary gslib files

```python
parameters = {
    'datafl' : '../pygslib/data/true_log.dat',        # path to file, or none (to use '_xxx_.in') or numpy array (with columns [x,y])
    'ivar'   : [1],                        # variables column numbers to be used in ivtail and ivhead,
    'tmin'   : -1000,                      # trimming limits min and max (raws out of this range will be ignored)
    'tmax'   : 1000,
    'igrid'   : 1,                          # 1=regular grid, 0=scattered values
    'nx':50,'ny':50,'nz':1,                 # if igrid=1: nx, ny, nz
    'xsiz':1,'ysiz':1,'zsiz':1,             # if igrid=1: xsiz, ysiz, zsiz
    'icolx':None,'icoly':None,'icolz':None, # if igrid=0: columns for x, y, z coordinates
    'outfl': None,                          # path to the output file or None (to use '_xxx_.out')
    'nxlag':30,'nylag':30,'nzlag':0,        # nxlag, nylag, nzlag
    'dxlag':1,'dylag':1,'dzlag':1,          # dxlag, dylag, dzlag
    'minpairs': 5,                          # minimum number of pairs
    'standardize': 1,                       # standardize sill? (0=no, 1=yes)
    'ivpar': [[1, 1, 1, None],
              [1, 1, 4, None]]}              # tail, head, variogram type, and cut (with shape [nvarg,4])    
    
```

```python
result, vtlist = varmap.varmap  (parameters)
```

```python
print(result.shape)
result.tail()
```

```python
pygslib.vtktools.SaveImageData(vtlist[0], 'vamap0')
pygslib.vtktools.SaveImageData(vtlist[1], 'vamap1')
```
